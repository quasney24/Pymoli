```python
import pandas as pd
import json
import os
```

```python
json_file = "purchase_data.json"

pymoli_pd = pd.read_json(json_file)

gaming_df = pd.DataFrame(pymoli_pd, columns =['SN','Age','Item ID','Item Name','Price'])
gaming_df.head()
```

```python
#Total players
length_players = len(gaming_df['SN'].value_counts())
df_total_players = pd.DataFrame({"Total Players": length_players}, index =[0])
df_total_players
```

# Purchasing Total

```python
#Purchasing Analysis

#Number of Unique Items 
df_unique_items = len(gaming_df['Item Name'].value_counts())
df_unique_items
#Average Purchase Price
df_average_price = gaming_df['Price'].mean()
df_average_price
#Total Number of Purchases
df_Total_Purchase = len(gaming_df)
df_Total_Purchase
#Total Revenue
df_total_revenue = gaming_df['Price'].sum()
df_total_revenue
#dataframe
purchasing_Total = pd.DataFrame({"Unique Items":[df_unique_items],"Average Price":[df_average_price],"Total Purchase":[df_Total_Purchase],"Total Revenue":[df_total_revenue]}, index = [0])

#print
purchasing_Total

#improvements.. change to 2 decimal pts .. dollar signs etc.. 
```

# Purchasing Analysis (Gender)

```python
#group by gender
gender_group = pymoli_pd.groupby(["Gender"])

#Purchase Count
tPurchase_count = gender_group["SN"].count()

#Average Purchase Price
taveragePrice = gender_group["Price"].mean()
#Total Purchase Value
tPurchase = gender_group ["Price"].sum()
#normalized Totals 
duplicate_data = pymoli_pd.drop_duplicates(subset='SN', keep="first")
grouped_duplicate_data = duplicate_data.groupby(["Gender"])

Norm = (gender_group["Price"].sum() / grouped_duplicate_data["SN"].count())


#dataframe
pAnalysis_df = pd.DataFrame({"Purchase Count":tPurchase_count, "Average Purchase Price": taveragePrice, "Total Purchase Value": tPurchase, "Normalized": Norm })
pAnalysis_df
```

```python
#Binning
bins = [0,10,15,20,25,30,35,40,200]
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bin_L = ['Under 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', 'Over 40']

#bin to DF 
bin_df = pymoli_pd.copy()
bin_df["Age Groups"] = pd.cut(bin_df["Age"], bins, labels =bin_L)
bin_C = pd.cut(bin_df["Age"], bins, labels = bin_L)
grouped_bin = bin_df.groupby(["Age Groups"])

bin_count= grouped_bin["SN"].count()
count_total = pymoli_pd["SN"].count()
percent = (bin_count/count_total) *100
percent

A_percent = pd.DataFrame({"Total Count": bin_count,
                           "Percentage of Players": percent})

A_percent["Percentage of Players"] = A_percent["Percentage of Players"].map("{:.2f}%".format)
A_percent.head()

#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Normalized Totals

```

```python
#Binning
bins = [0,10,15,20,25,30,35,40,200]
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bin_L = ['Under 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', 'Over 40']

#bin to DF 
bin_df = pymoli_pd.copy()
bin_df["Age Groups"] = pd.cut(bin_df["Age"], bins, labels =bin_L)
bin_C = pd.cut(bin_df["Age"], bins, labels = bin_L)
grouped_bin = bin_df.groupby(["Age Groups"])

count_p = grouped_bin["Age"].count()
Average_p = grouped_bin["Price"].mean()
Total_p = grouped_bin["Price"].sum()

#delete dups
duplicate = pymoli_pd.drop_duplicates(subset = 'SN', keep = "first")
duplicate["Age Groups"] = pd.cut(duplicate["Age"], bins, labels = bin_L)
duplicate = duplicate.groupby(["Age Groups"])

bin_normalize = (grouped_bin["Price"].sum()/duplicate["SN"].count())
bin_normalize

#dataframe
age_group_df = pd.DataFrame({"Purchase Count": count_p,
                            "Average Purchase Price": Average_p,
                            "Total Purchase Value": Total_p,
                            "Normalized Totals": bin_normalize})

age_group_df


```

```python
#Top Spenders
#group by SN
SN_group = pymoli_pd.groupby(["SN"])

SN_group_count = SN_group["Item ID"].count()
SN_group_total = SN_group["Price"].sum()
SN_group_Average = (SN_group_total/SN_group_count)

SN_df = pd.DataFrame({"Purchase Count": SN_group_count, "Average Purchase Price": SN_group_Average,"Total Purchase Value":SN_group_total})

SN_df = SN_df.sort_values("Total Purchase Value", ascending = False)
SN_df.head(5)
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value
```

```python
#Most Popular Items
item_grouped = pymoli_pd.groupby(["Item ID", "Item Name"])
item_grouped_SN = item_grouped["SN"].count()
price_grouped_sum = item_grouped["Price"].sum()
item_grouped_price = (price_grouped_sum/item_grouped_SN)
item_grouped_value = (item_grouped_price * item_grouped_SN)

most_popular_items = pd.DataFrame({"Purchase Count": item_grouped, 
                                   "Item Price": item_grouped_price,
                                   "Total Purchase Value":item_grouped_value})
most_popular_items
#most_popular_items = most_popular_items.sort_values("Purchase Count", ascending = False)
#most_popular_items
#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```
